```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup complete")
```

```python
fifa_filepath = "./fifa.csv"
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

### Line Charts

```python
# Set the width and height of the figure
plt.Figure(figsize=(14,6))

# Add title
plt.title("FIFA Data")
# Line chart showing how FIFA ranking evolved over time
sns.lineplot(data=fifa_data)
```

### Plot a subset of the data

```python
list(fifa_data.columns)
```

```python
# Set the width and height of the figure
plt.Figure(figsize=(14,6))
# Add  title
plt.title("Fifa graph")

sns.lineplot(data=fifa_data['BRA'], label="BRA label")
sns.lineplot(data=fifa_data['FRA'], label="FRA label")

plt.ylabel("Score")
plt.xlabel("Year")
```

### Bar Chart

```python
# Path of the file to read
flight_filepath = "./flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

### Heatmap

 For instance, if you look closely, the months toward the end of the year (especially months 9-11) appear relatively dark for all airlines. This suggests that airlines are better (on average) at keeping schedule during these months!

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

### Scatter Plots

```python
# Path of the file to read
insurance_filepath = "./insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```
