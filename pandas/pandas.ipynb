### How to read a CSV file using pandas

````python
wine_reviews = pd.read_csv("./winemag-data-130k-v2.csv")
````

````python
wine_reviews
````

### Head() grabs the first 5 rows

````python
wine_reviews.head()
````

### You can see one unnamed coloumn added by default while useing `pd.read_csv`. To make pandas use that column for the index (instead of creating a new one from scratch), we can specify an `index_col`.

````python
wine_reviews = pd.read_csv("./winemag-data-130k-v2.csv", index_col=0)
wine_reviews.head()
````

### To select the first row of data in a DataFrame

````python
wine_reviews.iloc[0]
````

### To get a column with iloc, we can do the following. (Both works in principal loc and iloc are row-first, column-second)

````python
wine_reviews.iloc[:,0]
````

### the : operator, which also comes from native Python, means "everything". 

````python
wine_reviews.iloc[:,2]
````

### Possible ways

````python
wine_reviews.iloc[1:3,0]
````

````python
wine_reviews.iloc[[0, 1, 2], 0]
````

````python
wine_reviews.iloc[-5:]
````

### Lable Based Selection - here below line means go to `0` index row and find `country` column value of that row

````python
wine_reviews.loc[0,'country']
````

### `iloc` selects data by row and column position (numbers), while `loc` selects data by row/column labels (names or index values)


````python
wine_reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
````

### Conditional selection - We can start by checking if each wine is Italian or not

````python
wine_reviews.country == 'Italy'
````

````python
wine_reviews.loc[wine_reviews.country == 'Italy']
````

### We can use the ampersand `&` to bring the two questions together

````python
wine_reviews.loc[(wine_reviews.country == 'Italy') & (wine_reviews.points >= 90)]
````

### Suppose we'll buy any wine that's made in Italy or which is rated above average. For this we use a pipe `|`

````python
wine_reviews.loc[(wine_reviews.country == 'Italy') | (wine_reviews.points >= 90)]
````

### Assigning data

````python
wine_reviews['country'] = 'everyone'
wine_reviews['country']
````

````python
wine_reviews['index_backwards'] = range(len(wine_reviews), 0, -1)
wine_reviews['index_backwards']
````

````python
wine_reviews.head()
````

### Summary functions

````python
wine_reviews.points.describe()
````

````python
wine_reviews.points.mean()
````

````python
wine_reviews.designation.describe()
````

### Groupwise analysis

### `groupby()` created a group of reviews which allotted the same point values to the given coloumn . `value_counts()` is just a shortcut to this `groupby()` operation

````python
wine_reviews.groupby('points').points.count()
````

### to get the cheapest wine in each point value category, we can do the following

````python
wine_reviews.groupby('points').price.min()
````

### Multi-indexes

````python
countries_reviewed = wine_reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
````

### Sorthing

````python

countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
````

````python
countries_reviewed.sort_values(by='len', ascending=False)
````

````python
countries_reviewed.sort_index()
````

### sort by more than one column at a time

````python
countries_reviewed.sort_values(by=['country', 'len'])
````

### Dtypes

### To check datatype of coloumn

````python
wine_reviews.price.dtype
````

````python
wine_reviews.dtypes
````

### It's possible to convert a column of one type into another wherever such a conversion makes sense by using the `astype()` function

````python
wine_reviews.points.astype('float64')
````

### Missing data

### Entries missing values are given the value `NaN`, short for Not a Number. For technical reasons these NaN values are always of the `float64` dtype.

````python
wine_reviews[pd.isnull(wine_reviews.taster_twitter_handle)]
````

### if we want to replace a non null value then we use `replace()`

````python
wine_reviews.taster_twitter_handle.replace("@kerinokeefe", "@kerino")
````

### Renaming

### `rename()`, which lets you change index names and/or column names

````python
wine_reviews.rename(columns={'points': 'score'})
````

````python
wine_reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
````

# Data Selection

## `iloc` (Position Based)

```python
df.iloc[0]          # First row
df.iloc[:,0]        # First column
df.iloc[1:3,0]      # Slice
df.iloc[-5:]        # Last 5 rows
```

## `loc` (Label Based)

```python
df.loc[0,"country"]
df.loc[:,["country","points"]]
```

### Difference

| `iloc` | `loc` |
|--------|-------|
| Uses positions | Uses labels/index names |

# Conditional Selection

Equal:

```python

```

Filter:

```python
df.loc[df.country == "Italy"]
```

AND (`&`)

```python
df.loc[
    (df.country=="Italy") &
    (df.points>=90)
]
```

OR (`|`)

```python
df.loc[
    (df.country=="Italy") |
    (df.points>=90)
]
```
